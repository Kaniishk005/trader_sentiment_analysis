```python
def datetime(a):
    stra = ""
    count = 0
    for i in a:
        stra+=i
        count+=1
        if(count==10):
            break
    return stra
df2['date'] = df2['Timestamp IST'].apply(datetime)
```

```python
df2['Timestamp IST'] = pd.to_datetime(df2['Timestamp IST'], dayfirst=True)
df2['date'] = pd.to_datetime(df2['date'], dayfirst=True)
df1['date'] = pd.to_datetime(df1['date'])
df2.head()
# Converted argument to date time format
```

```python
print("Sentiment range:")
print(df1['date'].min(), "to", df1['date'].max())

print("\nTrades range:")
print(df2['date'].min(), "to", df2['date'].max())
```

```python
# Lets merge data
df = pd.merge(df2,df1[['date','classification','value']],on = 'date',how = 'left')
df.head(10)
```

```python
df.isnull().sum()
```

```python
# Number of null values are less therefore we simply drom them
df = df.dropna(subset = 'classification')
df.isnull().sum()
```

```python
df['classification'].value_counts()
# Sentiment distribution
```

## 2. DATA ANALYSIS

```python
df['win'] = df['Closed PnL'] > 0
df[['Closed PnL','win']].head()
# Creating win column for columns with PnL(profit nd loss)>0
```

```python
# Lets divide or quantile cut data on basis size USD in trade size category
df['size_category'] = pd.qcut(df['Size USD'], q=3, labels=['Small','Medium','Large'])
df[['Size USD','size_category']].head()
```

```python
df['position_type'] = df['Side'].str.upper()
df['position_type'].value_counts()
```

```python
pnl = df.groupby(by = 'classification')['Closed PnL'].mean().sort_values()
pnl
```

```python
fig = plt.figure(figsize=(8,5))
sns.barplot(x = pnl.index,y = pnl.values,palette='rainbow')
plt.title("Mean pnl vs Sentiment")
plt.xlabel("Classification")
plt.ylabel("Mean Closing PnL")
plt.xticks(rotation=45)
```

```python
winSentiment = df.groupby('classification')['win'].mean()
winSentiment
```

```python
plt.figure(figsize=(8,5))

sns.barplot(
    x=winSentiment.index,
    y=winSentiment.values,
    palette='rainbow'
)

plt.title("Win Rate by Market Sentiment")
plt.xlabel("Market Sentiment")
plt.ylabel("Win Rate")

plt.xticks(rotation=45)

plt.show()
```

### Insight 1: Trader Profitability vs Market Sentiment

Trader profitability varies significantly across different market sentiment regimes.

Average PnL increases as market sentiment becomes more optimistic:

- Neutral: ~34
- Extreme Fear: ~34
- Greed: ~43
- Fear: ~54
- Extreme Greed: ~68

This suggests traders benefit from strong  momentum during 
Extreme Greed periods where price trends are clearer.

### Insight 2: Win Rate vs Market Sentiment

Win rate also varies by sentiment regime.

- Extreme Greed shows the highest win rate (~46%)
- Fear periods show relatively strong win rates (~42%)
- Extreme Fear has the lowest win rate (~37%)

This indicates that traders perform best during strong greed 
sentiment and struggle during extreme fear conditions.

```python
sentiment = pd.crosstab(df['classification'], df['position_type'])
sentiment
```

```python
sentiment_pct = sentiment.div(sentiment.sum(axis=1), axis=0)
sentiment_pct
```

```python
sentiment_pct.plot(
    kind='bar',
    stacked=True,
    figsize=(9,6)
)

plt.title("Long vs Short Behavior by Market Sentiment")
plt.xlabel("Market Sentiment")
plt.ylabel("Proportion of Trades")

plt.legend(title="Position Type")

plt.xticks(rotation=45)

plt.show()
```

### Insight 3: Long vs Short Behavior Across Sentiment

Trader positioning slightly shifts across sentiment regimes.

During Extreme Greed periods, a larger proportion of trades are SELL orders (~55%).  
This suggests traders may anticipate price corrections or take profits when the market becomes overheated.

During Extreme Fear periods, BUY trades slightly dominate (~51%), which could indicate traders attempting to buy the dip.

```python
size_sentiment = df.groupby('classification')['Size USD'].mean()

size_sentiment
```

```python
plt.figure(figsize=(8,5))

sns.barplot(
    x=size_sentiment.index,
    y=size_sentiment.values,
    palette = 'cubehelix'
)

plt.title("Average Trade Size by Market Sentiment")
plt.xlabel("Market Sentiment")
plt.ylabel("Average Trade Size (USD)")

plt.xticks(rotation=45)

plt.show()
```

### Insight 4: Trade Size vs Market Sentiment

Average trade size varies significantly across sentiment regimes.

The largest trades occur during **Fear periods (~7816 USD)**, suggesting traders may take larger positions when markets are uncertain, possibly attempting to capture rebounds.

In contrast, **Extreme Greed periods show the smallest trade sizes (~3112 USD)**, indicating traders may reduce risk exposure when markets appear overheated.

```python
trader_activity = df.groupby('Account').size().reset_index(name='trade_count')

trader_activity.head()
```

```python
threshold = trader_activity['trade_count'].median()

trader_activity['trader_type'] = np.where(
    trader_activity['trade_count'] > threshold,
    'Frequent Trader',
    'Infrequent Trader'
)

trader_activity.head()
```

```python
df = df.merge(
    trader_activity[['Account','trader_type']],
    on='Account',
    how='left'
)
```

```python
segment_performance = df.groupby('trader_type')['Closed PnL'].mean()

segment_performance
```

```python
plt.figure(figsize=(7,5))

sns.barplot(
    x=segment_performance.index,
    y=segment_performance.values
)

plt.title("Performance: Frequent vs Infrequent Traders")
plt.ylabel("Average PnL")

plt.show()
```

### Insight 5: Frequent vs Infrequent Traders

Frequent traders exhibit different performance characteristics compared to infrequent traders.

This suggests that trading frequency may influence profitability, potentially due to overtrading or better market timing among certain trader groups.

 ## 4. Bonus Model

```python
trader_features = df.groupby('Account').agg({'Closed PnL':'mean','Size USD':'mean','win':'mean'}).reset_index()
trader_features.head()
```

```python
X = trader_features[['Closed PnL','Size USD','win']]
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

X_scaled = scaler.fit_transform(X)
```

```python
# Applying K means
from sklearn.cluster import KMeans

kmeans = KMeans(n_clusters=3, random_state=42)

trader_features['cluster'] = kmeans.fit_predict(X_scaled)
```

```python
trader_features.head()
```

```python
plt.figure(figsize=(8,6))

sns.scatterplot(
    x=trader_features['Size USD'],
    y=trader_features['Closed PnL'],
    hue=trader_features['cluster'],
    palette='Set1'
)

plt.title("Trader Behavioral Clusters")
plt.xlabel("Average Trade Size")
plt.ylabel("Average PnL")

plt.show()
```

```python
cluster_summary = trader_features.groupby('cluster')[['Closed PnL','Size USD','win']].mean()

cluster_summary
```

### Bonus: Trader Behavioral Archetypes

K-Means clustering was applied to group traders based on average profitability,
average trade size, and win rate.

Three behavioral archetypes emerge:

**Cluster 0 – Aggressive traders**
- Large average trade sizes (~11k USD)
- Moderate profitability (~162 average PnL)
- Balanced win rate (~43%)

**Cluster 1 – Conservative traders**
- Smaller trade sizes (~3.3k USD)
- Lower profitability (~33 average PnL)
- Moderate win rate (~40%)

**Cluster 2 – High-risk / high-reward traders**
- Large trade sizes (~10.6k USD)
- Very high profitability (~470 average PnL)
- Lower win rate (~29%), suggesting fewer but larger winning trades

These clusters demonstrate how trader behavior and risk appetite vary across
the platform, highlighting distinct trading styles.
